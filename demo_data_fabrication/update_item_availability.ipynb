```python
# Get all the stores
url = "https://gateway-staging.ncrcloud.com/site/sites/find-by-criteria"

querystring = {"pageSize":"1000","pageNumber":"0"}

payload = "{\n  \"criteria\": {\n \n  }\n}"
headers = {
    'Content-Type': "application/json",
    'nep-application-key': "8a00860b6641a0ae0166471356ba000f",
    'Authorization': "Basic YWNjdDpqYW1AamFtc2VydmljZXVzZXI6MTIzNDU2Nzg=",
    'Cache-Control': "no-cache",
    'Postman-Token': "7e0e8b36-4cd7-405b-8d92-4cffb474bcd0"
    }

response = requests.request("POST", url, data=payload, headers=headers, params=querystring)
print (response.text)
resp_json = json.loads(response.text)
```

```python
print (resp_json["pageContent"][0]["id"])
```

```python
# Create availability

url = "https://gateway-staging.ncrcloud.com/ias/item-availability"

n_items = 1500

prob_av = 0.5




for i in range(len(resp_json["pageContent"])):
    siteId = resp_json["pageContent"][i]["id"]
    item_availability = [{"availableForSale": np.random.rand() < prob_av, "itemCode": str(j)} for j in range(n_items)]
    payload_dict = {"multiItemAvailabilityWriteData": item_availability}
    payload = json.dumps(payload_dict, indent=2)
    # print (payload)
    headers = {
        'nep-application-key': "8a00860b6641a0ae0166471356ba000f",
        'nep-enterprise-unit': siteId,
        'Content-Type': "application/json",
        'Authorization': "Basic YWNjdDpqYW1AamFtc2VydmljZXVzZXI6MTIzNDU2Nzg=",
        'Cache-Control': "no-cache",
        'Postman-Token': "00c1acff-09da-49c3-ab9a-d2b37bd3f69c"
    }
    response = requests.request("PUT", url, data=payload, headers=headers)
    print (response.text)
```
